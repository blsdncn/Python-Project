```python
setA = set([1,2,3,4,5]) #also {1,2,3,4,5}
setB = set([4,5,6,7,8]) #also {4,5,6,7,8}
print(1 in setA)
print(1 in setB)
```

Dictionaries:
- A set of key-value pairs
- Dictionaries are mutable

```python
d = {'a':1, 'b':2, 'c':3}
x = d['b']
print(x)

#Entries can be changed by assigning something to that entry.
d['b'] = 5
print(d['b'])

#Assigning to a key that doesn't exist adds an entry
d['d'] = 5
print(d)
```

<u><b>Python as a functional language</u></b>

Python supports functional programming concepts and features:

```python
numbers = [1,2,3,4,5]
#Using map ot apply a funtion to each element in the list
squared_numbers = list(map(lambda x: x**2,numbers))
print(squared_numbers)
#Using filter to select elements that satisfy a condition
even_numers = list(filter(lambda x : x % 2 == 0, numbers))
print(even_numers)
```

<u><b>Python as a logical language</u></b>

Python doesn't have built-in support for pattern matching, unification, and backtracking in the same way that languages like Prolog do. However, you can achieve similar functionality using libraries or by implementing your own logic.

```python
class FamilyTree:
    def __init__(self):
        self.facts = []

    def add_fact(self, fact):
        self.facts.append(fact)

    def query(self, pattern):
        for fact in self.facts:
            if self.match_pattern(pattern, fact):
                return True
        return False

    def match_pattern(self, pattern, data):
        return all(p == d or (isinstance(p, str) and p.startswith('?') and
    data.get(p[1:]) == d) for p, d in zip(pattern, data))

# Example usage:
family_tree = FamilyTree()
# Define some facts in the family tree
family_tree.add_fact(["John", "is_a", "father"])
family_tree.add_fact(["Jane", "is_a", "mother"])
family_tree.add_fact(["John", "has_child", "Alice"])
family_tree.add_fact(["Jane", "has_child", "Alice"])
family_tree.add_fact(["Alice", "is_a", "daughter"])
# Query the family tree
query_pattern = [
    ["?parent", "has_child", "?child"],
    ["?parent", "is_a", "father"]
]

if family_tree.query(query_pattern):
    print("Pattern found in the family tree!")
else:
    print("Pattern not found in the family tree.")
```
